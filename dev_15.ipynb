```python
from day_15 import Day15, Step


day = Day15("./input/day_15.txt")
grid1 = day.grid2
r, c = day.start2

def move(r: int, c: int, step: Step, grid):
    # we probably have to do something with moving a list of coordinates in the 
    # case we hit a box ("[]")
    dr, dc = step.delta
    nr , nc = r + dr, c + dc
    
    if grid[nr][nc] in ["[", "]"]:
        if step in [Step.LEFT, Step.RIGHT]:
            temp = move(nr, nc, step, grid)
        else:
            if grid[nr][nc] == "[":
                temp1 = move(nr,nc,step,grid)
                if temp1 == step.delta:
                    temp2 = move(nr,nc+1,step,grid)
                    if temp2 != step.delta:
                        grid[nr][nc], grid[nr+dr][nc+dc] = grid[nr+dr][nc+dc], grid[nr][nc]
            if grid[nr][nc] == "]":
                temp1 = move(nr,nc,step,grid)
                if temp1 == step.delta:
                    temp2 = move(nr,nc-1,step,grid)
                    if temp2 != step.delta:
                        grid[nr][nc], grid[nr+dr][nc+dc] = grid[nr+dr][nc+dc], grid[nr][nc]


    if grid[nr][nc] == ".":
        # swap the adjecent elements
        grid[r][c], grid[nr][nc] = grid[nr][nc], grid[r][c]
        return step.delta
    elif grid[nr][nc] == "#":
        return 0,0
        
    return 0,0


for step in day.moves:
    dr, dc = move(r, c, step, grid1)
    r += dr
    c += dc

ans = 0


for r, row in enumerate(grid1):
    for c, cell in enumerate(row):
        if cell == "[":
            ans += r * 100 + c


print(ans)
    


    
    



    

```

```python
from copy import deepcopy
day = Day15("./input/day_15.txt")
grid = day.grid2
r, c = day.start2

for move in day.moves:
    dr, dc = move.delta
    targets = [(r,c)]
    go = True
    for cr, cc in targets:
        nr = cr + dr
        nc = cc + dc
        if (nr,nc) in targets: continue
        char = grid[nr][nc]
        if char == "#":
            go = False
            break
        if char == "[":
            targets.append((nr,nc))
            targets.append((nr,nc+1))
        if char == "]":
            targets.append((nr,nc))
            targets.append((nr,nc-1))
    if not go: continue 
    copy = deepcopy(grid)
    grid[r][c] = "." 
    grid[r + dr][c + dc] = "@"
    for br, bc in targets[1:]:
        grid[br][bc] = "."
    for br, bc in targets[1:]:
        grid[br + dr][bc + dc] = copy[br][bc]
    r += dr
    c += dc

ans = 0

for r, row in enumerate(grid):
    for c, cell in enumerate(row):
        if cell == "[":
            ans += r * 100 + c

print(ans)
    
```
